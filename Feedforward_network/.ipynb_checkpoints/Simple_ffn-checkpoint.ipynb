```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### Buliding feedforward Neural network 

```python
Nclass = 500 # 500 samples per class
# create three cloud as normal distribution
X1 = np.random.randn(Nclass, 2)+np.array([0,-2])
X2 = np.random.randn(Nclass, 2)+np.array([2,2])
X3 = np.random.randn(Nclass, 2)+np.array([-2,2])

X = np.vstack([X1,X2,X3])
# create labels
Y = np.array([0]*Nclass + [1]*Nclass + [2]*Nclass)


```

```python
# plottig data
plt.scatter(X[:,0], X[:,1], c = Y, alpha = 0.6)
plt.show()
```

```python
# Setting paramenters for NN

D= 2 # input size of neuron
M= 3 #  hidden layer size of nueron
K = 3 # Output layer: there are 3 classes

# intialize the weights

W1 = np.random.randn(D,M) # Wight matrix for I/P layer
b1 = np.random.randn(M)   # bias at input layer
W2 = np.random.rand(M,K) # weight matrix at hd layer
b2 = np.random.rand(K) # bias at output
```

```python
# define sigmoid function
def sigmoid(x,deriv=False):
    if(deriv==True):
        return x*(1-x)
    return 1/(1+np.exp(-x))
```

```python
# define the forward acttion
def feed_forward(X, W1, b1, W2, b2):
    Z = sigmoid(X.dot(W1)+b1)
    A = Z.dot(W2)+b2
    expA = np.exp(A)
    Y = expA/expA.sum(axis = 1, keepdims = True) # probability
    return Y
```

```python
# defining classification rate

def classification_rate(Y, P):
    n_correct = 0
    n_total = 0
    for i in xrange(len(Y)):
        n_total += 1
        if Y[i] == P[i]:
            n_correct += 1
    return float(n_correct)/n_total


    
```

```python
# cal the forward function
P_Y_given_X = feed_forward(X, W1, b1, W2, b2)
P = np.argmax(P_Y_given_X, axis =1)
```

```python
assert(len(P) == len(Y))
```

```python
print "classifcation rate for randomy chosen weights"
print classification_rate(Y, P)
```
