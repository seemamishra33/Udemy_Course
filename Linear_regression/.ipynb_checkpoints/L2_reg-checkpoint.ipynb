```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
N = 50
X = np.linspace(0,10)
Y = 0.5*X+np.random.randn(N)  # Added some random noise

# Add some outlier

Y[-1]+= 30
Y[-2]+= 30

plt.scatter(X,Y)
plt.show()
```

```python
# maximum likelihood solution

X = np.vstack([np.ones(N), X]).T
w_ml = np.linalg.solve(X.T.dot(X), X.T.dot(Y))
Yhat_ml = X.dot(w_ml)
plt.scatter(X[:,1], Y)
plt.scatter(X[:,1], Yhat_ml)
plt.show()
```
